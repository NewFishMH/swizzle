```python
from preprocessing.funnel import Funnel
from preprocessing.prepro import PreProcessor
from postprocessing.postpro import PostProcessor

import numpy as np
import librosa
import pandas as pd
import plotly.express as px
import numpy as np
#import os
import math
import time
#import plotly.graph_objects as go
from tensorflow import keras  
```

```python
p = PreProcessor()
f = Funnel(preprocessor=p)
X = f.process_data('../data/raw/audio_hex-pickup_debleeded/05_Rock3-117-Bb_solo_hex_cln.wav')
X_bypass = np.load('../data/output/05_Rock3-117-Bb_solo_hex_cln_data.npz')['arr_0']

####### ACCESS PREPRO DIRECTLY ######

# create PreProcessor object
p1 = PreProcessor(verbose=4)

# get audio and label file for first filename
# audio, labels = p.load_files("05_Rock3-117-Bb_solo.jams", rec_mode='pd')

audio = '../data/raw/audio_hex-pickup_debleeded/05_Rock3-117-Bb_solo_hex_cln.wav'
audio, _ = librosa.load(audio, sr=None)

# Preprocess audio and labels
p1.preprocess_audio(data=audio, training=True)
X_direct = p1.output['windows']
# p.preprocess_labels(labels)


print(p.output['windows'])

swizzle_model = keras.models.load_model("../app/model/swizzle_model_1song_of_landscape", compile=False)

y_pred = swizzle_model.predict(X)
y_pred_bypass = swizzle_model.predict(X_bypass)
y_pred_direct = swizzle_model.predict(X_direct)


postpro = PostProcessor()
post_pro_output = postpro.postprocess_data(y_pred)
post_pro_output_bypass = postpro.postprocess_data(y_pred_bypass)
post_pro_output_direct = postpro.postprocess_data(y_pred_direct)
```
